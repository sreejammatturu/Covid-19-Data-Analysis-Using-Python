# COVID-19 Data Analysis Using Python

### Importing the Python Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Loading the Covid-19 Confirmed Cases Dataset

```python
df_covid = pd.read_csv('covid19_Confirmed_dataset.csv')
df_covid.head()
```

```python
df_covid.shape
```

Dropping the useless columns from the dataset

```python
#df_covid.drop(['Lat','Long'], axis = 1, inplace = True)
df_covid.head(15)
```

### Aggegating the Country/Region Column

```python
df_covid_agg = df_covid.groupby('Country/Region').sum()
df_covid_agg.tail(20)
```

```python
df_covid_agg.shape
```

Plotting the Covid-19 Cases for USA, China and Spain

```python
df_covid_agg.loc["US"].plot()
df_covid_agg.loc["Spain"].plot()
df_covid_agg.loc["China"].plot()

plt.legend()
```

### Calculating the Good Measure

```python
df_covid_agg.loc["China"][:3].plot()
```

Maximum Derivation Visualization for China

```python
df_covid_agg.loc["China"].diff().plot()
```

```python
df_covid_agg.loc["China"].diff().max()
```

```python
df_covid_agg.loc["Italy"].diff().max()
```

```python
df_covid_agg.loc["US"].diff().max()
```

### Finding the max infection rate for all countries and putting in dataset by creating a column called max_infect_rate

```python
countries = list(df_covid_agg.index)
max_infect_rate = []
for c in countries:
    max_infect_rate.append(df_covid_agg.loc[c].diff().max())
df_covid_agg["max_infect_rate"] = max_infect_rate
```

```python
df_covid_agg.head(10)
```

```python
df_covid_agg['max_infect_rate'].max()
```

### Creating a new dataframe with only needed column

```python
df_covid_data = pd.DataFrame(df_covid_agg["max_infect_rate"])
df_covid_data.head(10)
```

## Importing the World Happiness dataset

• Importing the WorldHappinessRepost.csv dataset<br>
   • Selecting needed columns for analysis.<br>
   • Join the datasets.<br>
   • Calculating correlations as the result of our analysis.

```python
df_happy = pd.read_csv('worldwide_happiness_report.csv')
df_happy.head(10)
```

```python
#Making the Country column as index
df_happy.set_index("Country or region", inplace = True)
```

```python
df_happy.head(10)
```

```python
df_happy.shape
```

Dropping the useless columns from the world happiness dataset

```python
useless_cols = ["Overall rank","Score","Generosity","Perceptions of corruption"]
```

```python
df_happy.drop(useless_cols, axis = 1, inplace = True)
```

```python
df_happy.shape
```

### Performing Inner Join to join the covid-19 dataset and world happiness dataset

```python
df_join = df_covid_data.join(df_happy, how = 'inner')
```

```python
df_join.head()
```

```python
df_join.shape
```

### Calculating Correlation Matrix between two datasets

```python
df_join.corr()
```
